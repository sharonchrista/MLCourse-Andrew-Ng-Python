```python
def feature_normalize(X):
    # Initialize variables
    X_norm = X.copy()
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)

    # Perform feature normalization
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma

# Example usage:
# X_norm, mu, sigma = feature_normalize(X)
```

```python
def warm_up_exercise():
    # Return a 5x5 identity matrix
    return np.eye(5)

def plot_data(x, y):
    plt.plot(x, y, 'rx', markersize=10)
    plt.xlabel('Population of City in 10,000s')
    plt.ylabel('Profit in $10,000s')
    plt.show()

def compute_cost(X, y, theta):
    m = len(y)
    J = (1 / (2 * m)) * np.sum((X.dot(theta) - y) ** 2)
    return J

def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros(num_iters)

    for iter in range(num_iters):
        error = X.dot(theta) - y
        theta -= (alpha / m) * X.T.dot(error)
        J_history[iter] = compute_cost(X, y, theta)

    return theta, J_history

def plot_cost_contour(X, y, theta_history):
    theta0_vals = np.linspace(-10, 10, 100)
    theta1_vals = np.linspace(-1, 4, 100)
    J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

    for i, theta0 in enumerate(theta0_vals):
        for j, theta1 in enumerate(theta1_vals):
            theta = np.array([theta0, theta1])
            J_vals[i, j] = compute_cost(X, y, theta)

    J_vals = J_vals.T

    # Surface plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(theta0_vals, theta1_vals, J_vals, cmap='viridis')
    ax.set_xlabel('\u03B8_0')
    ax.set_ylabel('\u03B8_1')
    ax.set_zlabel('Cost J')

    # Contour plot
    plt.figure()
    plt.contour(theta0_vals, theta1_vals, J_vals, levels=np.logspace(-2, 3, 20))
    plt.xlabel('\u03B8_0')
    plt.ylabel('\u03B8_1')
    plt.scatter(theta_history[:, 0], theta_history[:, 1], marker='x', s=70, c='r')
    plt.show()

def main():
    # Load data
    data = np.loadtxt('/content/drive/MyDrive/ex1data1.txt', delimiter=',')
    X = data[:, 0]
    y = data[:, 1]
    m = len(y)

    # Part 1: Warm-up exercise
    print('5x5 Identity Matrix:')
    print(warm_up_exercise())
    input('Press enter to continue.')

    # Part 2: Plotting
    print('Plotting Data...')
    plot_data(X, y)
    input('Press enter to continue.')

    # Part 3: Gradient Descent
    print('Running Gradient Descent...')
    X = np.column_stack((np.ones(m), X))  # Add a column of ones to X
    theta = np.zeros(2)
    iterations = 1500
    alpha = 0.01

    # Compute and display initial cost
    print('Initial Cost:', compute_cost(X, y, theta))

    # Run gradient descent
    theta, _ = gradient_descent(X, y, theta, alpha, iterations)

    # Print theta to screen
    print('Theta found by gradient descent:', theta)

    # Plot the linear fit
    plt.plot(X[:, 1], X.dot(theta), '-')
    plt.legend(['Training data', 'Linear regression'])
    plt.show()

    # Predict values for population sizes of 35,000 and 70,000
    predict1 = np.dot([1, 3.5], theta)
    print('For population = 35,000, we predict a profit of $', predict1 * 10000)
    predict2 = np.dot([1, 7], theta)
    print('For population = 70,000, we predict a profit of $', predict2 * 10000)

    input('Press enter to continue.')

    # Part 4: Visualizing J(theta_0, theta_1)
    print('Visualizing J(theta_0, theta_1)...')
    plot_cost_contour(X, y, theta_history=np.array([theta]))


if __name__ == "__main__":
    main()
```
